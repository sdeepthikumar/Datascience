```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import train_test_split
```

```python
feature = data.drop(columns='TARGET')
```

```python
target = data.TARGET
target.value_counts()
```

```python
x_train,y_train,x_test,y_test = train_test_split(feature,target)
```

```python
vt = VarianceThreshold(threshold=0) #removing constant features
```
