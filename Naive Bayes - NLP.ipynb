```python
horror_data = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/horror-train.csv')
```

```python
horror_data.info()
```

```python
horror_data.author.value_counts()
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
re_token = RegexpTokenizer(r'[A-Za-z]+')
```

```python
horror_data.text
```

```python
re_token.tokenize(horror_data.text[0])
```

```python
horror_data['tokenised_text']= horror_data.text.map(lambda t :re_token.tokenize(t))
```

```python
from nltk.stem.snowball import SnowballStemmer 
stemmer = SnowballStemmer('english')
```

```python
horror_data['stemmed_token'] = horror_data['tokenised_text'].map(lambda s:[stemmer.stem(eachword) for eachword in s])
horror_data['stemmed_token']
```

```python
horror_data
```

```python
horror_data['tokenised_sentence']= horror_data['stemmed_token'].map(lambda t:' '.join(t))
```

```python
horror_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
feature = cv.fit_transform(horror_data.tokenised_sentence)
feature.toarray()
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
from sklearn.model_selection import train_test_split
trainX, testX, trainY, testY = train_test_split(feature, horror_data.author)
mnb.fit(trainX,trainY)
```

```python
mnb.score(testX,testY)
```

```python
mnb.predict(testX[:1])
```

```python
testY[:1]
```

```python
y_pred = mnb.predict(testX)
```

```python
confusion_matrix(y_pred=y_pred, y_true=testY)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(trainX, trainY)
lr.score(testX, testY)
y_pred = lr.predict(testX)
confusion_matrix(y_pred=y_pred, y_true=testY)
```

```python
class MyNaive_Bayes:
    def __init__(self):
        self.likelihood_tables = {}
        
    def get_frequency_table(self,feature_data,target_data):
        freq_dict = {}
        for eachcol in feature_data.columns:
            frequency_table = pd.crosstab(feature_data[eachcol],tennis_data.play)
            freq_table_dict[eachcol] = frequency_table
        return freq_table_dict
    
    def get_likelihood_table(self,freq_table_dict):
        likelihood_table_dict = {}
        
        for col,freq_table in freq_table_dict.items():
            likelihood_table = freq_table
            
            for col_name in freq_table.columns:
                total_count = freq_table[col_name].sum()
                likelihood_table[col_name]= freq_table[col_name]/total_count
                likelihood_table_dict[col] = likelihood_table
        return likelihood_table_dict
        
    def myfit(self,feature_data,target_data):
        freq_table_dict = self.get_frequency_table(feature_data,target_data)
        likelihood_table = self.get_likelihood_table(freq_table_dict)
        self.likelihood_tables = likelihood_table
        
        target_freq = target_data.value_counts()
        target_events = target_data.value_counts().sum()
        self.class_prior_probability = target_freq/target_events
    
    def mypredict(self,data):
        for eachrow in data:
            print(eachrow)
            p_yes = 1
            for col,value in eachrow.items():
                p_yes *= self.likelihood_tables[col].loc[value]['yes']
                p_yes *= self.class_prior_probability['yes']
                
            p_no = 1
            for col,value in eachrow.items():
                p_no *= self.likelihood_tables[col].loc[value]['no']
                p_no *= self.class_prior_probability['no']
                
            yes = p_yes/(p_yes+p_no)
            no = p_no/(p_yes+p_no)
            if yes > no:
                print ('Yes')
            else:
                print ('No')
      
```

```python
mn = MyNaive_Bayes()
feature_data = tennis_data.drop(columns='play')
target_data = tennis_data.play

mn.myfit(feature_data,target_data)
```
